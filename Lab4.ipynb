```python
#Author: Caden LeCluyse
#Date: 2/23/24
#File Lab4.ipynb
#Purpose Implement heap sort

#This defines the function that will convert the list into a max heap
def maxHeapify(heapList: list, index: int, heapSize: int):
  #The left index is 2 times the index passed in plus one, the right index is simply the spot to the right
  left = 2*index + 1
  right = left + 1

  if left < heapSize and heapList[left] > heapList[index]: #If the left child is less than the heap size, and if the value of the left index is greater than the value passed in
    largest = left #we set the largest index to be the left index, otherwise we set the largest index to be the index that was passed in
  else:
    largest = index
  if right < heapSize and heapList[right] > heapList[largest]: #We simply do the same here with the right child
    largest = right
  if largest != index: #if the largest index isnt the index that was passed in, we swap the largest index and the index passed in, and then recursively call maxheapify with the largest index
    heapList[index], heapList[largest] = heapList[largest], heapList[index]
    maxHeapify(heapList, largest, heapSize)

def buildMaxHeap(heapList: list, size: int, heapSize: int):
  for i in range(size//2 - 1, -1, -1):
    maxHeapify(heapList, i, heapSize)

#Defines the heapsort function
def heapSort(heapList: list, size: int):
  heapSize = size #set the heapSize to be the size of the list, and then build the heap
  buildMaxHeap(heapList, size, heapSize)

  #Loop through the list starting at the end index to the beginning index
  for i in range(size - 1, 0, -1):
    heapList[0], heapList[i] = heapList[i], heapList[0] #swap the ith index with the 0th index
    heapSize -= 1 #Subtract one from the heap size
    maxHeapify(heapList, 0, heapSize) #return the sorted max heap
  return heapList

def tester():
  testList = [4, 1, 3, 2, 16, 9, 10, 14, 8, 7]
  print("Test case 1:", testList)
  print("Sorted list: ", heapSort(testList, len(testList)), "\n\n")

  testList = [789, 567, 425, 876, 4, 72, 16, 92, 18, 24, 1, -5, 16, 1024, 27, 257, 124, 99]
  print("Test case 2:", testList)
  print("Sorted list: ", heapSort(testList, len(testList)), "\n\n")


tester();
```
