```python
#Author: Caden LeCluyse
#Date: 3/29/24
#Filename: Lab7.ipynb
#Purpose: Implement a disjoint set

#Create a Node class
class Node:

  #Create the constructor for a node, that takes in an entry
  def __init__(self, entry):
    self.value = entry
    self.parent = self
    self.rank = 0


#Simply calls the Node constructor with the given entry
def make_set(entry):
  return Node(entry)

#Defines the function that will union two trees
def union(set_one: Node, set_two: Node):
  link(find_set(set_one), find_set(set_two))


def link(set_one: Node, set_two: Node):
  #If the first set has a bigger rank than the second,
  #We set the parent of the second set to be the first set
  if set_one.rank > set_two.rank:
    set_two.parent = set_one
 #otherwise, set the first set's parent to be the second set, and updates the rank
 #This is the only function where the rank will be increased
  else:
    set_one.parent = set_two
    if(set_one.rank == set_two.rank):
      set_two.rank = set_two.rank + 1

#If the parent isnt the root, the root becomes the parent
#otherwise return the parent
def find_set(entry: Node):
  if entry != entry.parent:
    entry.parent = find_set(entry.parent)
  return entry.parent


def main():
  #97 is 'a' and 104 is 'h'
  start_char = 97
  end_char = 104

  #Defines the disjoint set, which will be a list of nodes
  disjoint_set = []

  #This loops from 'a' to 'h', and creates a set based on the character
  #The set is then added to the list
  for i in range(start_char, end_char + 1):
    disjoint_set.append(make_set(chr(i)))

  #Do the unions that are required
  union(disjoint_set[1], disjoint_set[0])
  union(disjoint_set[3], disjoint_set[2])
  union(disjoint_set[5], disjoint_set[4])
  union(disjoint_set[5], disjoint_set[6])
  union(disjoint_set[2], disjoint_set[1])
  union(disjoint_set[6], disjoint_set[7])
  union(disjoint_set[3], disjoint_set[6])
  find_set(disjoint_set[5])

  #This will print the parent of each node!
  print("All parents: ")
  for entry in disjoint_set:
    print(entry.value + ".parent = " + entry.parent.value)

  #This will print the rank of each node
  print("\n\nAll ranks:")
  for entry in disjoint_set:
    print(str(entry.value) + ".rank = " + str(entry.rank))

main()
```
