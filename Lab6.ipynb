```python
#Author: Caden LeCluyse
#Date: 3/22/2024
#File: Lab6.ipynb
#Purpose: Implement a hashtable
#Note: I have already made a hashtable in c++ for a personal project
#Note: If I was doing this on my own I would've put the hash function in its own
#      function, but I didn't know if that was allowed for the assignment

#In this lab the given hash function that I am using will be stored with the table itself
class HashTable:
  #Initializes an empty array of size n and stores the size for the table
  #I am slowly learning python naming conventions
  def __init__(self, size: int, function: int):
    self._table = [None] * size
    self.table_size = size
    self.hash_function = function

  #This function will insert a key into the hashtable
  def insert_key(self, key: int):
    hash_value = 0
    #Get the hash value based on which hash function I want to use
    if self.hash_function == 1:
      #Define the subfunctions
      sub_function_one = key % 13
      sub_function_two = 1 + (key % 11)

      #loop through the list while also getting the index
      for index in range(self.table_size):
        #Do the hash function we were given to implement
        hash_value = (sub_function_one + (index * sub_function_two)) % self.table_size

        #If the hash index in the table is empty, set it equal to the key and break
        if self._table[hash_value] is None:
          self._table[hash_value] = key
          break
    elif self.hash_function == 2:
      #Define subfunctions
      sub_function_one = key % 7
      sub_function_two = 5 - (key % 5)

      #loop through the list while also getting the index
      for index in range(self.table_size):
        #Do the hash function we were given to implement
        hash_value = (sub_function_one + (index * sub_function_two)) % self.table_size

        #If the hash index in the table is empty, set it equal to the key and break
        if self._table[hash_value] is None:
          self._table[hash_value] = key
          break

  #This will find and return a key index pair
  def search_key(self, key: int):
    hash_value = 0
    if self.hash_function == 1:
      #Define subfunctions
      sub_function_one = key % 13
      sub_function_two = 1 + (key % 11)

      #loop through the list while also getting the index
      for index in range(self.table_size):
        #Do the hash function we were given to implement
        hash_value = (sub_function_one + (index * sub_function_two)) % self.table_size

        #If the hash index in the table is empty, we return a fail flag
        #Otherwise, simply return the hash value and the key
        if self._table[hash_value] is None:
          return (-1, key)
        elif self._table[hash_value] == key:
          return (hash_value, key)
    elif self.hash_function == 2:
      sub_function_one = key % 7
      sub_function_two = 5 - (key % 5)

      #loop through the list while also getting the index
      for index in range(self.table_size):
        #Do the hash function we were given to implement
        hash_value = (sub_function_one + (index * sub_function_two)) % self.table_size

        #If the hash index we got is empty in insert mode, return the hash value
        if self._table[hash_value] is None:
          return (-1, key)
        elif self._table[hash_value] == key:
          return (hash_value, key)
    return (-1, key)

  #A simple function to print out the hash table
  def print_table(self):
    print("[", end="")
    for index, value in enumerate(self._table):
      if value is not None:
          print(value, end="" if index == self.table_size - 1 else ", ")
      else:
          print("NIL", end="" if index == self.table_size - 1 else ", ")
    print("]\n")

def main():
  #Initialize both of the hash tables per assignment requirements
  hash_tables = [HashTable(13, 1), HashTable(7, 2)]

  #Create a for loop so I can do both tests without typing everything out again
  for index, table in enumerate(hash_tables):
    if index == 0:
      print("Test case one:")
    else:
      print("\n\n\nTest case two:")
    #Prompt the user to enter the list of numbers they want to put into the hash table
    list_input = input("\nPlease enter your list of numbers seperated by spaces: ")
    string_of_numbers = list_input.split()
    #Converts the string of characters into a list of ints
    list_to_insert = [int(char) for char in string_of_numbers]
    #Loop through the ints
    for key in list_to_insert:
      #Insert into the first or second hash table, respectively,
      #using the hash function indicated in the assignment guidelines
      table.insert_key(key)

    #Simple if else statement that determines if we are working with the first or second hash table
    if index == 0:
      #Print the given table
      print("Resulting hash table: ")
      table.print_table()
      #Get the search results from the assignment requirements
      search_results = [table.search_key(14), table.search_key(66)]

      #If the search didnt return -1, we print the index and value
      for search in search_results:
        if search[0] == -1:
          print(search[1], "could not be found in the hash table!")
        else:
          print("The key", search[1], "is in index", search[0])

    #Repeat for the second table
    else:
      print("Resulting hash table: ")
      table.print_table()
      search_results = [table.search_key(52), table.search_key(11)]
      for search in search_results:
        if search[0] == -1:
          print(search[1], "could not be found in the hash table!")
        else:
          print("The key", search[1], "is in index", search[0])



main()
```
